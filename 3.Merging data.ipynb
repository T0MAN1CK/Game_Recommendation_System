```python
import pandas as pd
```

```python
details_df = pd.read_csv(r'game_details_full.csv')
details_df
```

```python
games_df = pd.read_csv(r'steamgames_test.csv')
games_df
```

```python
matching_game_names = details_df['Game'].tolist()

games_df = games_df[games_df['Title'].isin(matching_game_names)]
games_df.reset_index(drop=True, inplace=True)

games_df
```

```python
print("Duplicates in games_df:", games_df['Title'].duplicated().sum())
print("Duplicates in details_df:", details_df['Game'].duplicated().sum())
```

```python
duplicates = games_df.duplicated(subset=['Title'], keep=False)

duplicate_rows = games_df[duplicates]

duplicate_rows
```

```python
games_df.drop_duplicates(subset=['Title'], keep='first', inplace=True)

details_df.drop_duplicates(subset=['Game'], keep='first', inplace=True)
```

```python
details_df.shape
```

```python
games_df.shape
```

```python
merged_df = pd.merge(games_df, details_df, left_on='Title', right_on='Game', how='inner')

# Create a new DataFrame with the desired columns
final_columns = ['Title', 'Original Price', 'Discounted Price', 'Release Date', 'Link',
                 'Game Description', 'Recent Reviews Summary', 'All Reviews Summary',
                 'Recent Reviews Number', 'All Reviews Number', 'Developer', 'Publisher',
                 'Supported Languages', 'Popular Tags', 'Game Features', 'Minimum Requirements']

merged_df = merged_df[final_columns]


merged_df.reset_index(drop=True, inplace=True)
merged_df
```

```python
merged_df.to_csv('merged_data.csv', index=False)
```
