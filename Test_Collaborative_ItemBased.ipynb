```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

column_names = ['User_ID', 'Game_Name', 'Action', 'Hours', 'To drop']
user_data = pd.read_csv('steam-200k.csv', header=None, names=column_names)
user_data = user_data.drop('To drop', axis=1)

matrix = user_data.pivot_table(index='User_ID', columns='Game_Name', values='Hours', fill_value=0)

item_similarity = cosine_similarity(matrix.T)  
item_similarity_df = pd.DataFrame(item_similarity, index=matrix.columns, columns=matrix.columns)

def get_item_based_recommendations(user_id, matrix, similarity_df, num_recommendations=10):

    played_games = matrix.loc[user_id]
    played_games = played_games[played_games > 0].index.tolist()

    scores = {}
    for game in played_games:
        for other_game, similarity in similarity_df[game].items():
            if other_game not in played_games:
                scores[other_game] = scores.get(other_game, 0) + similarity

    sorted_scores = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    recommendations = [game[0] for game in sorted_scores[:num_recommendations]]

    return recommendations

results = pd.DataFrame(columns=['User_ID', 'Recommendations'])

unique_users = user_data['User_ID'].unique()

for user in unique_users:
    user_recs = get_item_based_recommendations(user, matrix, item_similarity_df)
    results = results.append({'User_ID': user, 'Recommendations': user_recs}, ignore_index=True)

results.to_csv('user_recommendations_ItemBased.csv', index=False)
```
