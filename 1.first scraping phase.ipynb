```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
import time
```

```python
url = 'https://store.steampowered.com/search/results/?query&start=0&count=50&dynamic_data=&sort_by=_ASC&category1=998&supportedlang=english&snr=1_7_7_230_7&infinite=1'

def totalresults(url):
    r = requests.get(url)
    data = dict(r.json())
    totalresults = data['total_count']
    return int(totalresults)

print(totalresults(url))

def get_data(url):
    r = requests.get(url)
    if r.status_code == 200:
        data = dict(r.json())
        return data['results_html']
    else:
        print(f"Request failed with status code: {r.status_code}")
        return None

def parse_game(game):
    title = game.find('span', class_='title').text
    release_date = game.find('div', class_='search_released').text
    
    original_price_container = game.find('div', class_='discount_original_price')
    discounted_price_container = game.find('div', class_='discount_final_price')
    
    if discounted_price_container:
        discounted_price = discounted_price_container.text.strip()
        
        if original_price_container:
            original_price = original_price_container.text.strip()
        else:
            original_price = discounted_price  # Use discounted price as original price
    else:
        original_price = "Free"
        discounted_price = "Free"
    
    link = game['href']
    
    return {
        'Title': title,
        'Original Price': original_price,
        'Discounted Price': discounted_price,
        'Release Date': release_date,
        'Link': link
    }

def parse(data):
    soup = BeautifulSoup(data, 'html.parser')
    games = soup.find_all('a', class_='search_result_row')
    
    parsed_data = []
    for game in games:
        parsed_data.append(parse_game(game))
    
    return parsed_data

```

```python
def output(gameslist):
    gamesdf = pd.DataFrame(gameslist)
    gamesdf.to_csv('steamgames_test.csv', index=False)
    print("Data saved to steamgames_test.csv")

results = []

for x in range(0, totalresults(url), 50):
    data = get_data(f'https://store.steampowered.com/search/results/?query&start={x}&count=50&dynamic_data=&sort_by=_ASC&category1=998&supportedlang=english&snr=1_7_7_230_7&infinite=1')
    results.extend(parse(data))
    print("Results Scraped: ", x)
    time.sleep(1)

output(results)
```
